# xgboost

## 訓鍊

```
import xgboost as xgb
from sklearn.metrics import mean_squared_error
import pandas as pd
import numpy as np
```

```
from google.colab import drive
drive.mount('/content/drive')
```

```
train = pd.read_csv("/content/drive/My Drive/colab/Colab Notebooks/資料探勘作業一/AdultsDataSet_Finaltrain.csv")
train
```

```
# normalize
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
train1 = scaler.fit_transform(train)
train2 = pd.DataFrame(train1,columns=train.columns)
train2
```

```
train_X = train2.drop('hours_per_week', axis=1)
train_y = train2['hours_per_week']
```

```
data_dmatrix = xgb.DMatrix(data=train_X,label=train_y)
```

```
xg_reg = xgb.XGBRegressor(objective ='reg:linear', colsample_bytree = 0.3, learning_rate = 0.1,
                max_depth = 5, alpha = 10, n_estimators = 10)
```

```
xg_reg.fit(train_X,train_y)
```

```
preds_train = xg_reg.predict(train_X)
rmse = np.sqrt(mean_squared_error(train_y, preds_train))
print("train_RMSE: %f" % (rmse))
```

```
import scipy
import sklearn
from sklearn.model_selection import train_test_split
# from sklearn.metrics import mean_absolute_percentage_error
print('train_MAPE为：',mean_squared_error(train_y,preds_train))
```
