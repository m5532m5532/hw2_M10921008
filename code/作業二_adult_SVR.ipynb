```python
from sklearn import datasets
from sklearn.svm import LinearSVR
from sklearn.pipeline import make_pipeline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## 訓練

```python
df_train = pd.read_csv("C:/Users/jimmy/Downloads/專案作業二/AdultsDataSet_Finaltrain.csv")
df_train
```

```python
# normalize 正規畫
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_train1 = scaler.fit_transform(df_train)
df_train2 = pd.DataFrame(df_train1,columns=df_train.columns)
df_train2
```

```python
train_X = df_train2.drop('hours_per_week', axis=1) # 將hours per week 拿掉
train_y = df_train2['hours_per_week']
```

```python
clf = LinearSVR(random_state=0, tol=1e-5)
clf.fit(train_X,train_y)
```

```python
df_test = pd.read_csv("C:/Users/jimmy/Downloads/專案作業二/AdultsDataSet_Finaltest.csv")
```

## 測試

```python
# normalize
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_test1 = scaler.fit_transform(df_test)
df_test2 = pd.DataFrame(df_test1,columns=df_train2.columns)
df_test2
```

```python
df_test2_x = df_test2.drop('hours_per_week', axis=1)
df_test2_y = df_test2['hours_per_week']
```

```python
preds_test = clf.predict(df_test2_x)
```

```python
from sklearn.metrics import mean_squared_error
print('RMSE为：',np.sqrt(mean_squared_error(df_test2_y,preds_test)))
```

```python
import scipy
import sklearn
from sklearn.model_selection import train_test_split
# from sklearn.metrics import mean_absolute_percentage_error
print('MAPE为：',mean_squared_error(df_test2_y,preds_test))
```
